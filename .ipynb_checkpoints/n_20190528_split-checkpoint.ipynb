```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.signal import find_peaks_cwt, peak_prominences

from utils.kde import smooth
wl = [i for i in range(660, 921)]
%matplotlib inline
```

### 觀察穩定(30sec)

```python
data = np.loadtxt("data/raw/20190524_split/IJV/live/live_1_1.csv", delimiter=",")
calib_wl = np.loadtxt("data/raw/20190524_split/IJV/calib_wl.csv", delimiter=",")
```

```python
data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp
```

```python
print("檔案格式(秒, 波長數): ", data.shape)
```

### 確認各波長隨時間改變趨勢一樣

```python
plt.figure(figsize=(12, 6))
for i in range(740, 811, 10):
    plt.plot(data[:, i-660], label="{} nm".format(i))
plt.xlabel("time [sec]")
plt.ylabel("reflectance [gray scale]")
plt.grid()
plt.legend()
plt.xticks([i for i in range(0, 1201, 200)], labels=[str(int(i/20)) for i in range(0, 1201, 100)])
plt.show()
```

### 確認一樣後，直接計算平均

```python
plt.figure(figsize=(12, 6))
plt.plot(data.mean(1), label="average from 660 to 920nm")
plt.xlabel("time [sec]")
plt.ylabel("reflectance [gray scale]")
plt.grid()
plt.legend()
plt.xticks([i for i in range(0, 1201, 200)], labels=[str(int(i/20)) for i in range(0, 1201, 100)])
plt.show()
```

```python
date = "20190524"
idx = 6
data = np.loadtxt("data/raw/20190524_split/IJV/live/live_{}_1.csv".format(idx), delimiter=",")
calib_wl = np.loadtxt("data/raw/20190524_split/IJV/calib_wl.csv", delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b1_max = data[max_p, :].mean(0)
b1_min = data[min_p, :].mean(0)

plt.plot(b1_max)
plt.plot(b1_min)
plt.show()

data = np.loadtxt("data/raw/20190524_split/IJV/live/live_{}_2.csv".format(idx), delimiter=",")
# data = data[-600:]
calib_wl = np.loadtxt("data/raw/20190524_split/IJV/calib_wl.csv", delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b2_max = data[max_p, :].mean(0)
b2_min = data[min_p, :].mean(0)

plt.plot(b2_max)
plt.plot(b2_min)
plt.show()

data = np.loadtxt("data/raw/20190524_split/IJV/live/live_{}_3.csv".format(idx), delimiter=",")
calib_wl = np.loadtxt("data/raw/20190524_split/IJV/calib_wl.csv", delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b3_max = data[max_p, :].mean(0)
b3_min = data[min_p, :].mean(0)

plt.plot(b3_max)
plt.plot(b3_min)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(wl, -np.log(b1_max/b1_min), alpha=0.4, color="C0")
plt.plot(wl, smooth(-np.log(b1_max/b1_min), 3), color="C0", label="baseline")

plt.plot(wl, -np.log(b2_max/b2_min), alpha=0.4, color="C1")
# plt.plot(wl, smooth(-np.log(b2_max/b2_min), 3), color="C1", label="hyperventilation")
plt.plot(wl, smooth(-np.log(b2_max/b2_min), 3), color="C1", label="inhale")

# plt.plot(wl, -np.log(b3_max/b3_min), alpha=0.4, color="C2")
# plt.plot(wl, smooth(-np.log(b3_max/b3_min), 3), color="C2", label="rest")

plt.grid()
plt.legend()
plt.xlabel("wavelength [nm]")
plt.ylabel("calculated absorption coefficient [-]")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
base = 660
s_time = 600
e_time = 1800
total_time = 4801

wl_select = [760, 800, 850]
wl_select = [i-base for i in wl_select]
for i in wl_select:
    plt.plot(data[:, i], label="{} nm".format(i+base))
    
    
for text, time in zip(["start", "end"], [s_time, e_time]):
    plt.plot(
        [time, time], 
        [data[:, wl_select].min()*0.99, data[:, wl_select].max()*1.01], 
        color="C0",
        linestyle=":",
        linewidth=2
    )

    plt.text(time+90, data[wl_select].max()*1.01, text, fontsize=12)

    
plt.xticks([i for i in range(0, total_time, 1200)], labels=[str(int(i/1200)) for i in range(0, total_time, 1200)])
plt.grid()
plt.legend()

plt.xlabel("time [min]")
plt.ylabel("reflectance [gray scale]")
plt.show()
```
