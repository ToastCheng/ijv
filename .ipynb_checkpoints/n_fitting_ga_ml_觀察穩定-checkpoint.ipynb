## 20190524

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
for i in range(1, 6):
    df = pd.read_csv("fitting/20190524_split/1_{}/log/log.csv".format(i))
    print(df["i_s"].iloc[0])
    df[::-1].reset_index()["i_s"].plot(ax=ax)
ax.grid()
ax.set_ylim(0, 1)
ax.legend([str(i) for i in range(1, 6)])
ax.set_xlabel("gene index [-]")
ax.set_ylabel("ScvO2 [-]")
plt.show()
```
