```python
def rgb(r, g, b):
    r = hex(r)[2:]
    g = hex(g)[2:]
    b = hex(b)[2:]
    
    if len(r)==1:
        r = "0" + r
    if len(g)==1:
        g = "0" + g 
    if len(b)==1:
        b = "0" + b
        
    return "#{}{}{}".format(
        r, g, b)
rgb(160, 160, 200)
```

```python
plt.figure(figsize=(12, 6))
colors = [rgb(220, i, i) for i in range(10, 230, 20)]
for c, i in zip(colors, np.arange(0, 1.1, 0.1)):
    plt.plot(
        df["wavelength"],
        df["oxy"] * i + df["deoxy"] * (1-i), 
        label="Saturation: {:2d}%".format(int(100*i)),
        color=c
    )
plt.xlabel("wavelength [nm]")
plt.ylabel("absorption coefficient [1/cm]")
plt.legend()
plt.grid()
plt.show()

```
