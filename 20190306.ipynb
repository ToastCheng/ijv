```python
small = np.load("output/20190225_small/result/result.pkl")
large = np.load("output/20190121_large/result/result.pkl")
```

```python
wl = [i for i in range(650, 1001, 10)]
sds = ["20mm", "24mm", "28mm"]
```

```python
for i in range(3):
    plt.plot(wl, smooth(small[0, 70, i, :], 0.7), label="small: %s" % sds[i])
    plt.plot(wl, smooth(large[0, 70, i, :], 0.7), label="large: %s" % sds[i])    
    plt.grid()
    plt.legend()
    plt.show()
```

```python

for s in range(3):
    fig = plt.figure(figsize=(8, 6))
    for i in range(100):
        color_idx = i // 10
        if i % 10 == 0:
            plt.plot(wl, small[0, i, 0, :], color="C%d" % color_idx, label="%d%%-%d%%" % (i, i+10))
        else:
            plt.plot(wl, small[0, i, 0, :], color="C%d" % color_idx)
    plt.grid()
    plt.legend()
    plt.title("small, sds: " + sds[s])
    plt.xlabel("wavelength [nm]")
    plt.ylabel("reflectance [-]")
    plt.show()
    
```

```python
for s in range(3):
    fig = plt.figure(figsize=(8, 6))
    for i in range(100):
        color_idx = i // 10
        if i % 10 == 0:
            plt.plot(wl, large[0, i, 0, :], color="C%d" % color_idx, label="%d%%-%d%%" % (i, i+10))
        else:
            plt.plot(wl, large[0, i, 0, :], color="C%d" % color_idx)
    plt.grid()
    plt.legend()
    plt.title("large, sds: " + sds[s])
    plt.xlabel("wavelength [nm]")
    plt.ylabel("reflectance [-]")
    plt.show()
```
