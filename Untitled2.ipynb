```python
wl = [i for i in range(650, 1001, 10)]
```

```python
p_data = []
for p in _p_data:
    p_data.append(np.interp(wl, p_wl, p))
p_data = np.asarray(p_data)
```

```python
s_data = sim.iloc[:, 1:].values.T
```

```python
s_data.shape
```

```python
def stand(x):
    x = (x - x.mean())/x.std()
    return x
```

```python
pp_data = stand(p_data)
ss_data = stand(s_data)

for i in range(36):
    plt.title("wavelength" + str(wl[i]))
    for j in range(4):
        
        plt.scatter(pp_data[j, i], ss_data[j, i])
#     plt.legend()
    plt.grid()
    plt.show()
```

```python
for i in range(36):
    plt.title("wavelength: " + str(wl[i]))
    plt.scatter([1,2,3,4], stand(p_data[:, i]))
    plt.scatter([1,2,3,4], stand(s_data[:, i]))
    plt.plot([1,2,3,4], stand(p_data[:, i]), label="phantom")
    plt.plot([1,2,3,4], stand(s_data[:, i]), label="sim")
    plt.xticks([1,2,3,4], ["C", "H", "I", "K"])
    plt.legend()
    plt.grid()
    plt.show()
```
