```python
wl = [i for i in range(650, 1001, 10)]
phantom_interp = []
live_interp = []

for p in ["PC", "PH", "PI", "PK"]:
    phantom_interp.append(
        np.interp(wl, df["Wavelength (nm)"].tolist(), phantom[p].tolist())
    )
for l in ["cy1", "cy2"]:
    live_interp.append(
        np.interp(wl, df["Wavelength (nm)"], live[l])
    )

phantom_interp = np.asarray(phantom_interp)
live_interp = np.asarray(live_interp)
```
