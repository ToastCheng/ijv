```python
from glob import glob
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
wl = np.loadtxt("data/raw/20190609/IJV/calib_wl.csv", delimiter=",")
wl.shape
```

```python
path = "data/raw/20190609/IJV/live/live_*_*.csv"
ls = glob(path)
ls.sort(key=lambda x: (int(x.split('live_')[-1][0]), int(x.split('live_')[-1][2])))
ls
```

```python
spec = []
for l in ls:
    spec += [np.loadtxt(l, delimiter=",")]
spec[10] = spec[10][-50:]
spec[10].shape
```

```python
plt.figure(figsize=(16, 10))
ratio = ["832: 100%", "832: 70%", "832: 50%", "832: 30%", "832: 0%"]
for i, s in enumerate(spec):
    if i % 3 != 2:
        continue
    plt.plot(wl[250:-300], s.mean(0)[250:-300], label=ratio[i//3])
    plt.grid()
    plt.legend()
    plt.xlabel("wavelength [nm]")
    plt.ylabel("reflectance [grayscale]")
plt.show()
```
