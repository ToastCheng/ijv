```python
import numpy as np 
import matplotlib.pyplot as plt
from utils.kde import KDE

kde = KDE()
%matplotlib inline
```

```python
d_large = np.load("output/20190121_large/result/result.pkl")
d_small = np.load("output/20190121_small/result/result.pkl")
```

```python
d_large = d_large[0]
d_small = d_small[0]
wl = [i for i in range(650, 1001, 10)]
sds = [20, 24, 28]
```

```python
for s in range(d_large.shape[1]):
    plt.plot(wl, kde.fit(d_large[95, s, :], 0.8), label="large")
    plt.plot(wl, kde.fit(d_small[95, s, :], 0.8), label="small")
    plt.legend()
    plt.grid()
    plt.title("SDS: %dmm" % sds[s])
    plt.xlabel("wavelength[nm]")
    plt.ylabel("reflectance[-]")
    plt.show()
```
