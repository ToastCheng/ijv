```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.signal import find_peaks_cwt, peak_prominences

from utils.kde import smooth
wl = [i for i in range(660, 921)]
%matplotlib inline
```

```python
date = "20190528"
idx = 2
data = np.loadtxt("data/raw/{}_split/IJV/live/live_{}_1.csv".format(date, idx), delimiter=",")
calib_wl = np.loadtxt("data/raw/{}_split/IJV/calib_wl.csv".format(date), delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b1_max = data[max_p, :].mean(0)
b1_min = data[min_p, :].mean(0)

plt.plot(b1_max)
plt.plot(b1_min)
plt.show()

data = np.loadtxt("data/raw/{}_split/IJV/live/live_{}_2.csv".format(date, idx), delimiter=",")
# data = data[-600:]
calib_wl = np.loadtxt("data/raw/{}_split/IJV/calib_wl.csv".format(date), delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b2_max = data[max_p, :].mean(0)
b2_min = data[min_p, :].mean(0)

plt.plot(b2_max)
plt.plot(b2_min)
plt.show()

data = np.loadtxt("data/raw/{}_split/IJV/live/live_{}_3.csv".format(date, idx), delimiter=",")
calib_wl = np.loadtxt("data/raw/{}_split/IJV/calib_wl.csv".format(date), delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b3_max = data[max_p, :].mean(0)
b3_min = data[min_p, :].mean(0)

plt.plot(b3_max)
plt.plot(b3_min)
plt.show()

data = np.loadtxt("data/raw/{}_split/IJV/live/live_{}_4.csv".format(date, idx), delimiter=",")
calib_wl = np.loadtxt("data/raw/{}_split/IJV/calib_wl.csv".format(date), delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b4_max = data[max_p, :].mean(0)
b4_min = data[min_p, :].mean(0)

plt.plot(b3_max)
plt.plot(b3_min)
plt.show()

data = np.loadtxt("data/raw/{}_split/IJV/live/live_{}_5.csv".format(date, idx), delimiter=",")
calib_wl = np.loadtxt("data/raw/{}_split/IJV/calib_wl.csv".format(date), delimiter=",")

data_interp = []
for i in range(data.shape[0]):
     data_interp += [np.interp(wl, calib_wl, data[i, :])]
data_interp = np.asarray(data_interp)
data = data_interp

max_p = find_peaks_cwt(data.mean(1), np.arange(1, 20))
min_p = find_peaks_cwt(-data.mean(1), np.arange(1, 20))

plt.plot(data.mean(1))
plt.scatter(max_p, data.mean(1)[max_p])
plt.scatter(min_p, data.mean(1)[min_p])
plt.show()

b5_max = data[max_p, :].mean(0)
b5_min = data[min_p, :].mean(0)

plt.plot(b3_max)
plt.plot(b3_min)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(wl, -np.log(b1_max/b1_min), alpha=0.4, color="C0")
plt.plot(wl, smooth(-np.log(b1_max/b1_min), 3), color="C0", label="baseline")

plt.plot(wl, -np.log(b2_max/b2_min), alpha=0.4, color="C1")
plt.plot(wl, smooth(-np.log(b2_max/b2_min), 3), color="C1", label="hyperventilation")
# plt.plot(wl, smooth(-np.log(b2_max/b2_min), 3), color="C1", label="inhale")

# plt.plot(wl, -np.log(b3_max/b3_min), alpha=0.4, color="C2")
# plt.plot(wl, smooth(-np.log(b3_max/b3_min), 3), color="C2", label="rest")

plt.grid()
plt.legend()
plt.xlabel("wavelength [nm]")
plt.ylabel("calculated absorption coefficient [-]")
plt.show()
```
