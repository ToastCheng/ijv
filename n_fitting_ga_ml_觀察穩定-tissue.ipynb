```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# fig, ax = plt.subplots(1, 1, figsize=(12, 6))
# ax2 = ax.twinx()

for i in range(3, 4):
    fig, ax = plt.subplots(1, 1, figsize=(12, 6))
    ax2 = ax.twinx()
    df = pd.read_csv("fitting/phantom2/2_{}/log/log_180.csv".format(i))
    print(df[["spec_rmsp", "absorb_rmsp", "i_s"]].iloc[0:10])
    df[::-1].reset_index()["i_s"].plot(ax=ax) #, color="#99bb99")
#     df[::-1].reset_index()["s_w"].plot(ax=ax, color="#99bb44")
#     df[::-1].reset_index()["s_f"].plot(ax=ax, color="#44bbee")
    np.log(df[::-1]).reset_index()["spec_rmsp"].plot(ax=ax2, color="C1", linestyle='--')
    
    ax.grid()
    ax.set_ylim(0, 1)
    ax.legend([str(i) for i in range(1, 6)])
    ax.set_xlabel("gene index [-]")
    ax.set_ylabel("ScvO2 [-]")

    ax2.legend(["error" + str(i) for i in range(1, 6)])
    ax2.set_ylabel("rmsp [-]")


    plt.show()
```
