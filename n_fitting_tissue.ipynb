```python
from engine import Engine
from glob import glob 
from random import uniform
from copy import deepcopy

from utils.ga import GA
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def rmse(x, y):

    x, y = np.array(x), np.array(y)
    return 100 * np.sqrt(((x-y)**2)).mean()/y.mean()
```

```python
spec_list = glob("data/calibrated/20190619/IJV/20190619_*.csv")
spec_list.sort(key=
    lambda x: int(x.split('_')[-1].strip('.csv')) 
)
```

```python
geo_max=[2, 2.25, 5.21, 13, 6, 16, 3]
geo_min=[2, 2.25, 3.69, 13, 6, 16, 3]
log_phantom = []
```

```python
wl = [i for i in range(660, 811, 10)]
ratio = [1, 0.66, 0.5, 0.33, 0]
```

```python
for s, r in zip(spec_list, ratio):
    print(s)
    spec = pd.read_csv(s)
    spec = spec[
        (spec["wavelength"] >= wl[0])&
        (spec["wavelength"] <= wl[-1])
    ]
    target_max = spec["max"].values.T
    target_min = spec["min"].values.T
    
    ga = GA(5000, 24, target_max, target_min, geo_max, geo_min, iteration=500, phantom=True, ink_ratio=r)
    ga(plot=True, verbose=True)
    log_phantom += [deepcopy(ga)]
```
