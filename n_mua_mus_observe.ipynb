## 各種散射

```python
df["musp"] 
mus500 = 5.17
bmie = 1.0
g = 0.94

wl = np.asarray([i for i in range(660, 921)])

musp = mus500 * ((wl/500) ** (-bmie))
mus_wrong = musp/g
mus = musp/(1-g)
```

```python
plt.plot(wl, mus_wrong)
plt.plot(wl, mus)
```
