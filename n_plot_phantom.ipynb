```python
import pandas as pd 
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
df = pd.read_csv("data/processed/20190510/IJV/phantom/20190510.csv")
df = df[["wavelength"]+list("chik")]
df.head()
```

```python
measured750 = (df[df["wavelength"]==750].iloc[:, 1:].values - df[df["wavelength"]==740].iloc[:, 1:].values)/\
df[df["wavelength"]==740].iloc[:, 1:].values
```

```python
measured730 = (df[df["wavelength"]==730].iloc[:, 1:].values - df[df["wavelength"]==740].iloc[:, 1:].values)/\
df[df["wavelength"]==740].iloc[:, 1:].values
```

```python
df = pd.read_csv("CHIKEN/sim_20190522_no_prism.csv")
wl = [i for i in range(660, 921, 10)]
```

```python
sim750 = (df[df["wl"]==750].iloc[:, 1:].values - df[df["wl"]==740].iloc[:, 1:].values)/\
df[df["wl"]==740].iloc[:, 1:].values
```

```python
sim730 = (df[df["wl"]==730].iloc[:, 1:].values - df[df["wl"]==740].iloc[:, 1:].values)/\
df[df["wl"]==740].iloc[:, 1:].values
```

```python
plt.plot(measured750[0], label="measured")
plt.plot(sim750[0], label="simulated")
plt.legend()
plt.grid()
plt.title("750nm")
plt.xlabel("phantom")
plt.ylabel("ratio")
plt.xticks([i for i in range(4)], list("CHIK"))
```

```python
plt.plot(measured730[0], label="measured")
plt.plot(sim730[0], label="simulated")
plt.legend()
plt.grid()
plt.title("730nm")
plt.xlabel("phantom")
plt.ylabel("ratio")
plt.xticks([i for i in range(4)], list("CHIK"))
```

```python
for pid in "chik":
    plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)][pid], label="c")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["c"], label="c")
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["h"], label="h")
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["i"], label="i")
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["k"], label="k")
plt.legend()
plt.grid()
plt.xticks(wl)
plt.savefig("CHIKEN/sim_20190522_no_prism.png")
```
