```python
import pandas as pd 
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
df = pd.read_csv("CHIKEN/20190509_sim_chik.csv")
wl = [i for i in range(660, 921, 10)]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["c"], label="c")
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["h"], label="h")
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["i"], label="i")
plt.plot(wl, df[(df["wl"] > 650) & (df["wl"] < 930)]["k"], label="k")
plt.legend()
plt.grid()
plt.xticks(wl)
plt.savefig("CHIKEN/20190509_sim.png")
```
