```python
from glob import glob 
import json
```

```python
files = glob("fitting/*")
files.sort(key=lambda x: (
    int(x.split('/')[1].split('_')[0]),
    int(x.split('_')[-2]), 
    int(x.split('_')[-1].strip('.json'))
))
for f in files:
    with open(f, 'r') as ff:
        j = json.load(ff)
        print(f)
        print("scvO2: {:.2f}".format(j["IJV"]["ScvO2"]))
        print("rmsp: {:.4f}".format(j["rmsp"]))
```
