```python

def smooth(x, band_width=3):
	def gaussian(x, mean, std):

		x = np.asarray(x)
		denominator = std*np.sqrt(2*np.pi)
		numerator = np.exp(-0.5*((x-mean)/std)**2)
		distribution =  numerator/denominator
		distribution /= distribution.sum()

		return distribution
	
	smoothed = np.zeros(len(x))
	for idx, xx in enumerate(x):
		smoothed += xx*gaussian(np.arange(len(x)), idx, band_width)
	
	return smoothed 
```

```python
order = 7
x = np.array([1, 4, 3, 6, 3, 2, -1])
result = np.polyfit(np.array([i for i in range(len(x))]), x, order)
def reconstruct(x, param):
    result = 0
    for idx, p in enumerate(reversed(param)):
        result += p * (np.arange(0, x.shape[0]) ** idx)
    return result
print(result)
plt.plot(x)
plt.plot(reconstruct(x, result))
```
