```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0.1, 10, 0.1)

def w(x):
    y = 0
    for xx in x:
        y += np.exp(-xx)
    return y

cv = []
cv1 = []
cv2 = []
cv3 = []
cv4 = []

for xx in x:
    tmp1 = xx + xx * 0.05 * (2*np.random.rand(100)-1)
    
    cv += [np.std(tmp1)/np.mean(tmp1)]

    after = []
    for i in range(100):
        tmp = xx + xx * 0.05 * (2*np.random.rand(5)-1)
        after += [w(tmp)]
    cv1 += [np.std(after)/np.mean(after)]
    
    after = []
    for i in range(100):
        tmp = xx + xx * 0.05 * (2*np.random.rand(10)-1)
        after += [w(tmp)]
    cv2 += [np.std(after)/np.mean(after)]
    
    after = []
    for i in range(100):
        tmp = xx + xx * 0.05 * (2*np.random.rand(100)-1)
        after += [w(tmp)]
    cv3 += [np.std(after)/np.mean(after)]
    
    after = []
    for i in range(100):
        tmp = xx + xx * 0.05 * (2*np.random.rand(1000)-1)
        after += [w(tmp)]    
    cv4 += [np.std(after)/np.mean(after)]
    
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x, cv, label="input series")
plt.plot(x, cv1, label="output series (x5)")
plt.plot(x, cv2, label="output series (x10)")
plt.plot(x, cv3, label="output series (x100)")
plt.plot(x, cv4, label="output series (x1000)")
plt.legend()
plt.grid()
plt.xlabel("input size")
plt.ylabel("coefficient of variation")
plt.show()
```
