```python
import pandas as pd 
import numpy as np 
import os
```

```python
mua = pd.read_csv("CHIKEN/mua_in_mm.csv")
mus = pd.read_csv("CHIKEN/musp_in_mm.csv")
wl = [i for i in range(660, 921, 10)]
```

```python
mua_ = {}
mua_['c'] = np.interp(wl, list(mua["wl"]), list(mua["C"])) * 10
mua_['h'] = np.interp(wl, list(mua["wl"]), list(mua["H"])) * 10
mua_['i'] = np.interp(wl, list(mua["wl"]), list(mua["I"])) * 10
mua_['k'] = np.interp(wl, list(mua["wl"]), list(mua["K"])) * 10
```

```python
mus_ = {}
mus_['c'] = np.interp(wl, list(mus["wl"]), list(mus["C"])) * 10
mus_['h'] = np.interp(wl, list(mus["wl"]), list(mus["H"])) * 10
mus_['i'] = np.interp(wl, list(mus["wl"]), list(mus["I"])) * 10
mus_['k'] = np.interp(wl, list(mus["wl"]), list(mus["K"])) * 10
```

```python
for pid in "chik":
    path = os.path.join("mcml_phantom/input/{}.txt".format(pid))
    with open(path, "w+") as f:
        for a, s in zip(mua_[pid], mus_[pid]):
            f.write("{}\t{}\n".format(a, s))
```
